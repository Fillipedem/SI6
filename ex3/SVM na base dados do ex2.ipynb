```python
import numpy as np
import pandas as pd

# Base de Dados
from sklearn.model_selection import train_test_split
# Classificador
from sklearn.svm import SVC
# Metricas
from sklearn.metrics import accuracy_score
```

### Lendo dados

```python
X = pd.read_csv("./data/X_res", header=None)
y = pd.read_csv("./data/y", header=None)
```

```python
X.head(5)
```

```python
y.head(5) 
```

### Divindo base de treino/teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7)
```

### Classificador

```python
cs = [100, 25, 10, 5, 2, 1.0, 0.1, 0.01]
clfs = []

for c in cs:
    clf = SVC(C=c, kernel='linear')
    clf.fit(X_train, list(y_train[0]))
    
    clfs.append((clf, c))
```

### Testes

```python
X_train.head(2)
```

```python
for clf, c in clfs:
    y_pred = clf.predict(X_test)

    print("Score: ", accuracy_score(y_test, y_pred), " C: ", c)
```
